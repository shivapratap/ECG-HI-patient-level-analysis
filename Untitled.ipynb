```python
import numpy as np
import pandas as pd
```

```python
file_path = '2min_12_Patients_35Features.csv'
data = pd.read_csv(file_path)

# Labels
data['label'] = data['label'].map({"before_HI": 0, "during_HI": 1})
data['subject_id'] = data['File_Name'].apply(lambda x: str(x).split('waveform_')[0])
```

```python
data.to_csv("processed_data2.csv")
```

```python
# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(file_path)

# ============================================================
# CREATE PATIENT ID
# ============================================================

df["Patient_ID"] = (
    df["File_Name"]
    .astype(str)
    .str.extract(r'^(\d+)')
)

# ============================================================
# REMOVE INCOMPLETE PATIENT
# ============================================================

df = df[df["Patient_ID"] != "98182"].copy()

print("Patients:", df["Patient_ID"].nunique())

# ============================================================
# ENCODE LABELS
# ============================================================

df["y"] = df["label"].map({
    "before_HI": 0,
    "during_HI": 1
})

# ============================================================
# FEATURE LIST
# ============================================================

exclude_cols = [
    "Sl. No:",
    "Segment_Number",
    "File_Name",
    "Patient_ID",
    "label",
    "y"
]

feature_cols = [
    c for c in df.columns
    if c not in exclude_cols
]

print("Features:", len(feature_cols))
```

```python
df.to_csv("processed_data2.csv")
```
